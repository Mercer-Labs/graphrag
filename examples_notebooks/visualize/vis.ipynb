```python
import pandas as pd

data1 = {'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie'], 'City': [1, 2, 3]}
df1 = pd.DataFrame(data1)

data2 = {'ID': [1, 3, 4], 'City': ['New York', 'London', 'Paris']}
df2 = pd.DataFrame(data2)

# Inner join on the 'ID' column
merged_df = df1.merge(df2, on='ID', how='inner', suffixes=('_left', '_right'))
print(merged_df)
```
